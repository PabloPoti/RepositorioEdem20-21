```python
import requests
import json
url ="https://petstore.swagger.io/v2/pet/1"
response = requests.get(url)
json_data=json.loads(response.content)
print(json_data)
```

```python
import requests
import json
url ="https://petstore.swagger.io/v2/pet/findByStatus?status=available"
response = requests.get(url)
json_data=json.loads(response.content)
print(response.status_code)
print(json_data)
```

```python
data=json.loads("""{
  "id": 0,
  "category": {
    "id": 0,
    "name": "pastor aleman"
  },
  "name": "Chispas",
  "photoUrls": [
    "string"
  ],
  "tags": [
    {
      "id": 0,
      "name": "puppy"
    }
  ],
  "status": "available"
}""")
response = requests.post(url="https://petstore.swagger.io/pet/",json=data)
print(response.status_code)
print(response.content)
```
